```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
 
movies = pd.read_csv('E:\Agile\Python Data Science\P4-Movie-Ratings.csv')
movies.columns = ['Film', 'Genre', 'CriticsRating', 'AudienceRating',
       'BudgetMillions', 'Year']


import warnings
warnings.filterwarnings('ignore')

movies
```

```python
sns.set_style("darkgrid")
vis1 = sns.lmplot(data=movies , x = 'CriticsRating' , y = 'AudienceRating' , \
                 fit_reg=False , hue = 'Genre', size =8  )
```

### Working with KDE

```python
vis2 = sns.kdeplot(movies.CriticsRating, movies.AudienceRating)
```

```python
vis3 = sns.kdeplot(movies.CriticsRating, movies.AudienceRating  ,shade=True)
```

```python
vis3 = sns.kdeplot(movies.CriticsRating, movies.AudienceRating  ,shade=True , shade_lowest=False , cmap="Reds")
```

### Working with Subplots

```python
vis4 = sns.kdeplot(movies.BudgetMillions , movies.CriticsRating ,color='Green')
```

```python
vis5= sns.kdeplot(movies.BudgetMillions , movies.AudienceRating , color='Blue')
```

```python
f , ax = plt.subplots(1,2)           # the (1,2) are the number of rows and columns
vis4 = sns.kdeplot(movies.BudgetMillions , movies.CriticsRating ,color='Green')  
vis5= sns.kdeplot(movies.BudgetMillions , movies.AudienceRating , color='Blue')
```

```python
import warnings
warnings.filterwarnings('ignore')

f , axes = plt.subplots(1,2 , figsize=(12,6))           # the (1,2) are the number of rows and columns ,figszie makes the plots larger
vis4 = sns.kdeplot(movies.BudgetMillions , movies.CriticsRating ,color='Green' , ax=axes[0])  #using the ax=axes[]  
vis5= sns.kdeplot(movies.BudgetMillions , movies.AudienceRating , color='Blue' , ax=axes[1])  #we can specify the plot locations.
```

```python
f , axes = plt.subplots(1,2 , figsize=(12,6))           

vis4 = sns.kdeplot(movies.BudgetMillions , movies.CriticsRating ,color='Green' , ax=axes[0])    
vis5= sns.kdeplot(movies.BudgetMillions , movies.AudienceRating , color='Blue' , ax=axes[1])  

vis4.set(xlim=(-30,220))        #we limited the x range that viz4 shows us
```

```python
f , axes = plt.subplots(1,2 , figsize=(12,6) , sharex=True)           # now we share the x axis specifications with the two plots

vis4 = sns.kdeplot(movies.BudgetMillions , movies.CriticsRating ,color='Green' , ax=axes[0])   
vis5= sns.kdeplot(movies.BudgetMillions , movies.AudienceRating , color='Blue' , ax=axes[1])  

vis4.set(xlim=(-30,220))
```

```python
f , axes = plt.subplots(1,2 , figsize=(12,6) , sharex=True , sharey=True)           # now we share the y axis specifications with the two plots

vis4 = sns.kdeplot(movies.BudgetMillions , movies.CriticsRating ,color='Green' , ax=axes[0])   
vis5= sns.kdeplot(movies.BudgetMillions , movies.AudienceRating , color='Blue' , ax=axes[1])  

vis4.set(xlim=(-30,220))
```

```python
sns.set_style('darkgrid')
f , axes = plt.subplots(2,2 , figsize=(12,12))           # now we share the y axis specifications with the two plots


v1 = sns.kdeplot(movies.BudgetMillions , movies.CriticsRating ,color='Green' , ax=axes[0,0])   
v2= sns.kdeplot(movies.BudgetMillions , movies.AudienceRating , color='Blue' , ax=axes[0,1])  
v3= sns.violinplot(data=movies[movies.Genre=='Drama'], x= 'Year' , y='CriticsRating',palette='cubehelix' , ax=axes[1,0])
v4= sns.kdeplot(movies.CriticsRating , movies.AudienceRating , color='Purple', shade=True,shade_lowest=False, cmap='Blues_r'  , ax=axes[1,1])
v4= sns.kdeplot(movies.CriticsRating , movies.AudienceRating , cmap='gist_gray'  , ax=axes[1,1])

v1.set(xlim=(-30,220))
v2.set(xlim=(-30,220))

plt.show()
```
