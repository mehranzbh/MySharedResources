```python
# Defining function to plot different player stats
def myplot (data,playerlist =  ["KobeBryant","JoeJohnson","LeBronJames","CarmeloAnthony"]): # when a pramater is given a value, it means the default values are the ones that specified if the user/code does not mention it in the function.   
    Col = {"KobeBryant":'Black',"JoeJohnson":'Red',"LeBronJames":'Green',"CarmeloAnthony":'Blue',"DwightHoward":'Orange',"ChrisBosh":'Yellow',"ChrisPaul":'Purple',"KevinDurant":'Pink',"DerrickRose":'Aqua',"DwayneWade":'Peru'}
    Mrk = {"KobeBryant":'h',"JoeJohnson":'d',"LeBronJames":'o',"CarmeloAnthony":'P',"DwightHoward":'*',"ChrisBosh":'s',"ChrisPaul":'8',"KevinDurant":'^',"DerrickRose":'v',"DwayneWade":'o'}
    
    for name in playerlist:
        plt.plot(data[Pdict[name]] , c = Col[name] , ls = ':' , marker=Mrk[name] ,label = name)
    
    plt.xticks(list(range(0,10)),Seasons , rotation='vertical')
    plt.legend(loc='upper left' , bbox_to_anchor=(1,1)) # in the bbox tuple (0,0)  means beginning of x axis and y axis, so "upper right" of the legend is gonna be in the lower(0) left(0).
    plt.show()

myplot (Points , ["DwightHoward","ChrisBosh","ChrisPaul","KevinDurant","DerrickRose","DwayneWade"])
```
