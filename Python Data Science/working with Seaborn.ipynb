```python
import pandas as pd
stats = pd.read_csv('E:\Agile\Python Data Science\P4-Demographic-Data.csv')
stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers',
       'IncomeGroup']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#plt.rcParams['figure.figsize'] = 12,2
vis1 = sns.displot(stats["InternetUsers"],bins=10)
```

```python
vis2= sns.lmplot(data=stats, x="BirthRate" , y="InternetUsers" , \
    fit_reg=False)
```

```python
vis2= sns.lmplot(data=stats, x="BirthRate" , y="InternetUsers" , \
    fit_reg=False,hue="IncomeGroup")
```

```python
vis2= sns.lmplot(data=stats, x="BirthRate" , y="InternetUsers" , \
    fit_reg=False,hue="IncomeGroup" , size=10)
```

```python
vis2= sns.lmplot(data=stats, x="BirthRate" , y="InternetUsers" , \
    fit_reg=False,hue="IncomeGroup" , size=7 , aspect = 1.5)
```

```python
vis2= sns.lmplot(data=stats, x="BirthRate" , y="InternetUsers" , \
    fit_reg=False,hue="IncomeGroup" , size=5 , aspect = 1.5 ,    \
    scatter_kws= {"s":100}) #the scatter_kws passes the dictionary values to the 
                           #scatter plot in "scatter plot at matplotlib" and in the documentation
                           # the "s" stands for size
```
