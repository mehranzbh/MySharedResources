```python
#initiation

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

movies = pd.read_csv('E:\Agile\Python Data Science\P4-Movie-Ratings.csv')
movies.columns = ['Film', 'Genre', 'CriticsRating', 'AudienceRating',
       'BudgetMillions', 'Year']
       
```

```python
g = sns.FacetGrid(data=movies,  row='Genre' , col='Year' , hue='Genre')     #we specify row and column and color
```

```python
g = sns.FacetGrid(data=movies,  row='Genre' , col='Year' , hue='Genre')     #we should include both lines together so that the code to be executed
g = g.map(plt.scatter, 'CriticsRating' , 'AudienceRating')          #we specify the plot and x and y axis and we map it to the facet grid
```

```python
g = sns.FacetGrid(data=movies,  row='Genre' , col='Year' , hue='Genre')
g = g.map(plt.hist, 'BudgetMillions')          #we can specify any plot we want and pass the 
```

```python
g = sns.FacetGrid(data=movies,  row='Genre' , col='Year' , hue='Genre')   

kws= dict(s=50 , linewidth=0.5  , edgecolor = 'black')
kws

g = g.map(plt.scatter, 'CriticsRating' , 'AudienceRating' , **kws)      # we passed a new argument which is a dictionary 
                                                                        # and we customized the plots
```

### drawing diogonals

```python
g = sns.FacetGrid(data=movies,  row='Genre' , col='Year' , hue='Genre')   
kws= dict(s=50 , linewidth=0.5  , edgecolor = 'black')
g = g.map(plt.scatter, 'CriticsRating' , 'AudienceRating' , **kws) 

g.set(xlim=(0,100), ylim=(0,100))      # limiting the both axis to 0 to 100

for ax in g.axes.flat:              #drawing diagonal for each ax
    ax.plot((0,100),(0,100) , c='gray' , ls= '--')  # setting color and line style

g.add_legend()      #adds legend for the Genre
```
